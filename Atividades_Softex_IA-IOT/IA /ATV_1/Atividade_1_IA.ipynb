```python
# Importanto bibliotecas
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import metrics

import pandas as pd
```

Primeira árvore

```python
url = "https://raw.githubusercontent.com/Daniellineto/Atividades_Softex_IA-IOT/refs/heads/main/IA/ATV_1/iris.data"
dataset = pd.read_csv(url, header=None)
columns = len(dataset.columns)

y = dataset.iloc[:, -1]
X = dataset.iloc[:, :-1]
print(dataset)
```

```python
# Divisão 80% treino e 20% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=None, stratify=y)
```

```python
# Cria classificador e treina o modelo
model = tree.DecisionTreeClassifier(criterion="entropy")
model = model.fit(X_train, y_train)
```

```python
# Utiliza do modelo para fazer a previsão dos dados X_test
result = model.predict(X_test)

# Estimativa da precisão dos acertos
acc = metrics.accuracy_score(result, y_test)
show = round(acc * 100)
print("{}%".format(show))

# Comparação das respostas
print(list(result))
print(list(y_test))
```

```python
tree.plot_tree(model,filled=True)
```
